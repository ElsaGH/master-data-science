```python
# CHUNKS

sc = pd.read_csv('searches.csv.bz2', sep='^', usecols=['Destination', 'Date'], chunksize=1000000)
# s = pd.read_csv('searches.csv.bz2', sep='^', usecols=['Destination', 'Date'], nrows=100000)

all_chunks = pd.DataFrame()

for i, s_chunk in enumerate(sc):
    print(i)
    s=s_chunk.dropna()
    s=s[s.Destination.isin(['MAD','BCN','AGP'])] 
    s['Plot_date']=s.Date.str[0:7]
    grouped = s.groupby(['Plot_date','Destination']).count()
    grouped = grouped.reset_index()
    all_chunks = all_chunks.append(grouped)

all_chunks = all_chunks.groupby(['Plot_date','Destination']).sum().reset_index()
all_chunks.head(10)
   

to_plot=all_chunks.pivot('Plot_date','Destination','Date')
to_plot.plot()
```
